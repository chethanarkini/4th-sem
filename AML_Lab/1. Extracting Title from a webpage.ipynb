 # Program 1 to extract title from web page


```python
import requests
from bs4 import BeautifulSoup
```

```python
# Example 1: Extract Google title
url = "https://www.google.com"
response = requests.get(url) # Get a response and store it 
soup = BeautifulSoup(response.text, "html.parser")# The text attribute of the response object
                                                  # Beautifulsoap object will parse the html content
title = soup.find("title").text # assigns the text content of the <title> tag to the variable title.
print(title)
```

```python
# Example 2: https://nmamit.nitte.edu.in/ # Extract title from this
url="https://nmamit.nitte.edu.in/"
response=requests.get(url)
soup=BeautifulSoup(response.text, "html.parser")
title=soup.find("title").text
title
```

# 2. Extracting Links from the webpage

```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://nmamit.nitte.edu.in/"
url2="https://www.google.com"
response=requests.get(url)
response1=requests.get(url2)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
soup1= BeautifulSoup(response1.text, "html.parser")
```

```python
# nitte links
for link in soup.find_all("a"):  # soup.find_all("a") returns a list of all the <a> tags in the parsed HTML content represented by the soup object.
    print(link.get("href")) # link.get("href") extracts the value of the href attribute of the current <a> tag being iterated over, which is the URL of the hyperlink.
```

```python
# google links
for link in soup.find_all("a"):
    print(link.get("href"))
```

# 3. Extracting images from the webpage
